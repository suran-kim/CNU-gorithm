```python
n = int(input())
d = [0] * (n + 1) # 계산 횟수를 저장해주기 위해 만든다(인덱스가 2인 요소, 즉 2는 1로 만드려면 몇 번 계산해야 하는지, 인덱스가 3인 요소, 즉 3은 1로 만드려면 몇 번 계산해야 하는지......)
for i in range(2, n + 1): # 1은 애초에 1인 상태이므로 계산이 필요가 없다. (계산 횟수 0)
    d[i] = d[i - 1] + 1 # 일단 바로 직전의 수를 1로 만드는데 필요한 계산횟수에 1을 더해준다.
    # 위에서 1을 더해준다는 의미는, 1을 빼주는 연산을 사용했다는 의미이다. 
    if i % 3 == 0:
        d[i] = min(d[i], d[i // 3] + 1) # 1을 빼주는 방식이 계산 횟수가 더 작은지, 3으로 나누는 방식이 계산 횟수가 더 작은지 비교한다.
    if i % 2 == 0:
        d[i] = min(d[i], d[i // 2] + 1) # 1을 빼주는 방식이 계산 횟수가 더 작은지, 2로 나누는 방식이 계산 횟수가 더 작은지 비교한다. 
    # 여기서는 elif를 사용하면 안된다. 
    # elif를 사용하면 2로도 나누어지고 3으로도 나누어지는 수에 대해서 고려가 완전히 이루어지지 않는다. 
print(d[n])
```

# 풀이 

## 1. dp를 활용하는 문제이다. 바로 직전의 수, 그러니까 자신보다 1 작은 수에 대한 계산횟수를 미리 계산해두고, 그 값을 바로 꺼내 사용하는 것이다.
## 2. 1을 빼주는 연산은 3으로 나누는 연산/ 2로 나누는 연산과 코드 표현을 다르게 해준다. 어차피 현재 수에 1을 빼주고 나면 현재 수보다 1이 작은 수가 되고, 현재 수보다 1이 작은 수에 대한 계산횟수는 계산이 되어있다. 그러므로 이미 계산된 계산횟수에 1만 더해주면, 1을 빼주는 연산을 고려해준 것이나 다름없다.
## 3. 1을 빼는 연산이 좋은지, 2로 나누는 연산이 좋은지, 3으로 나누는 연산이 좋은지는 어떤 수이냐에 따라 달라진다. 그러므로, 세가지 연산을 모두 고려한다. 
## 4. if 문의 min에서 비교하는 것은 결국, 1을 빼주는 방식이 효율적인가 아니면 3이나 2로 나누어 떨어지는 수라면 3이나 2로 나누는 것이 효율적인가를 비교하는 것이다. 여기서 d[i // 3] + 1이나 d[i // 2] + 1에 1을 더해주는 이유는 현재 수를 3이나 2로 나눈 수의 계산횟수에서 3이나 2로 나누는 연산 횟수가 하나 더 추가되었기 때문이다. d[i]에 1을 더하지 않은 이유는, 윗 줄에서 이미 1을 더해주었기 때문이다.
## 5. 결국 핵심은 이것이다. 현재 수의 바로 직전의 수에 대한 최적의 계산횟수를 계산해 두고, 이 결과값을 바로바로 활용하여 현재 수를 1로 만드는 연산을 끝까지 하는 것을 방지하는 것. 
## 6. 3이나 2로 나누어 떨어지지 않으면, 어쩔 수 없이 1을 빼야 한다. 이때의 최적의 연산 횟수도 역시 '바로 직전의 수의 계산횟수+1'이다.
