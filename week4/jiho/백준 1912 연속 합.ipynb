```python
#  sys.stdin.readline  . (   )

n = int(input())
a = list(map(int, input().split())) # split()      + map int    + list ([])  

for i in range(1,n):
    a[i] = max(a[i], a[i-1]+a[i])
    
print(max(a))
```
