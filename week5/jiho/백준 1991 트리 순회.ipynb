## 풀이

```python
n = int(input())
inputs = []
for _ in range(n):
    inputs.append(input().split())

class Node(): # Node 클래스 생성 
    def __init__(self, item, left, right): # -> 해당 노드의 값과 왼쪽/오른쪽의 값을 인스턴스 변수로 갖게한다
        self.item = item
        self.left = left
        self.right = right

def preorder(node): # 전위 순회 
    print(node.item, end = '') # 공백 없이 출력(root)
    if node.left != '.': # 왼쪽 값 존재 시
        preorder(tree[node.left]) # 왼쪽값 출력
    if node.right != '.': # 오른쪽 값 존재 시 
        preorder(tree[node.right]) # 오른쪽 값 출력
        # root - left - right
        
def inorder(node):
    if node.left != '.': # 왼쪽 값 존재시
        inorder(tree[node.left]) # 왼쪽값 출력
    print(node.item, end = '') # 공백 없이 출력
    if node.right != '.':
        inorder(tree[node.right])
        # left - root - right
def postorder(node):
    if node.left != '.':
        postorder(tree[node.left])
    if node.right != '.':
        postorder(tree[node.right])
    print(node.item, end = '') # 공백 없이 출력
    # left - right - root

tree = {} # 딕셔너리 생성
for item, left, right in inputs:
    tree[item] = Node(item, left, right)

preorder(tree['A'])
print()
inorder(tree['A'])
print()
postorder(tree['A'])
```
